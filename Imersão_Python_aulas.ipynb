# **Aula 2 - Importação**


```python
import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel("/content/PuraImersão Python - Tabela de ações.xlsx", sheet_name="Principal")
df_principal
```

```python
df_Total_de_acoes = pd.read_excel("/content/PuraImersão Python - Tabela de ações.xlsx", sheet_name="Total_de_acoes")
df_Total_de_acoes
```

```python
df_Ticker = pd.read_excel("/content/PuraImersão Python - Tabela de ações.xlsx", sheet_name="Ticker")
df_Ticker
```

```python
df_Tabela_GPT = pd.read_excel("/content/PuraImersão Python - Tabela de ações.xlsx", sheet_name="Tabela_GPT")
df_Tabela_GPT
```

# **Aula 3 - Manipulação**

```python
df_principal = df_principal[['Ativo',	'Data',	'Último (R$)',	'Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Último (R$)':'valor_final','Var. Dia (%)':'var_dia_pct'}).copy()
df_principal
```

```python
df_principal["var_pct"] = df_principal["var_dia_pct"]/100
df_principal["valor_inicial"] = df_principal["valor_final"]/(df_principal["var_pct"]+1)
df_principal
```

```python
df_principal = df_principal.merge(df_Total_de_acoes, left_on="Ativo",right_on="Código",how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns=["Código"])
df_principal
```

```python
df_principal["variacao_rs"] = (df_principal["valor_final"] - df_principal["valor_inicial"])*df_principal["Qtde. Teórica"]
df_principal
```

```python
pd.options.display.float_format = '{:.2f}'.format
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica':'qtd_teorica'}).copy()
df_principal
```

```python
df_principal["resultado"] = df_principal['variacao_rs'].apply(lambda x: "Subiu" if x > 0 else ("Desceu" if x < 0 else "Estável"))
df_principal
```

```python
df_principal = df_principal.merge(df_Ticker, left_on="Ativo",right_on="Ticker",how='left')
df_principal = df_principal.drop(columns=["Ticker"])
df_principal
```

```python
df_principal = df_principal.merge(df_Tabela_GPT, left_on="Nome",right_on="Nome da Empresa",how='left')
df_principal = df_principal.drop(columns=["Nome da Empresa"])
df_principal
```

```python
df_principal = df_principal.rename(columns={'Idade (em anos)':'idade'}).copy()
df_principal
```

```python
df_principal["cat_idade"] = df_principal['idade'].apply(lambda x: "Mais de 100 anos" if x > 100 else ("Menos de 50 anos" if x < 50 else "Entre 50 e 100 anos"))
df_principal
```

```python
maior = df_principal["variacao_rs"].max()
menor = df_principal["variacao_rs"].min()
media = df_principal["variacao_rs"].mean()
media_subiu = df_principal[df_principal["resultado"] == "Subiu"]["variacao_rs"].mean()
media_desceu = df_principal[df_principal["resultado"] == "Desceu"]["variacao_rs"].mean()

print(f"Maior\tR$ {maior:,.2f}")
print(f"Menor\tR$ {menor:,.2f}")
print(f"Média\tR$ {media:,.2f}")
print(f"Média de quem subiu\tR$ {media_subiu:,.2f}")
print(f"Média de quem desceu\tR$ {media_desceu:,.2f}")
```

```python
df_principal_subiu = df_principal[df_principal["resultado"] == "Subiu"]
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby("Segmento")['variacao_rs'].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby("resultado")['variacao_rs'].sum().reset_index()
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x='resultado', y='variacao_rs', text='variacao_rs', title='Variações Reais por Resultado')
fig.show()
```

# **Desafio**

```python
fig.update_traces(texttemplate='R$ %{text:,.2f}', textposition='outside')
fig.show()
```

```python
fig = px.pie(df_analise_segmento, names='Segmento', values='variacao_rs', title='Variações Reais por Segmentos')
fig.show()
```

```python
segmentos_excluir = ['Petróleo e Gás', 'Mineração', 'Energia Elétrica', 'Bancário']
df_analise_segmento_filtrado = df_analise_segmento[~df_analise_segmento['Segmento'].isin(segmentos_excluir)]

fig = px.pie(df_analise_segmento_filtrado, values='variacao_rs', names='Segmento',
             title='Distribuição das Variações por Segmento')
fig.show()
```

```python
df_analise_faixa_etaria = df_principal.groupby('cat_idade').agg({'variacao_rs': 'sum', 'Ativo': 'count'}).reset_index()
df_analise_faixa_etaria
```

```python
df_analise_faixa_etaria.columns = ['Analise_faixa_etaria', 'Variacao_rs', 'Qtd_de_empresas']
df_analise_faixa_etaria
```

```python
fig = px.bar(df_analise_faixa_etaria, x='Analise_faixa_etaria', y='Variacao_rs',
             text='Variacao_rs', title='Variação de Receita por Faixa Etária')

fig.update_traces(texttemplate='R$ %{text:,.2f}', textposition='outside')
fig.show()
```
